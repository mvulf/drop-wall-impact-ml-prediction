# Logit

```python
estimator = StatsModelsEstimator(Logit)

ml_pipe = MLPipeline(
    target='splashing',
    estimator=estimator,
    model_postfix=model_postfix,
    add_smote=add_smote,
)

ml_pipe.run()
```

No fragmentation. without SMOTE

```python
estimator = StatsModelsEstimator(Logit)

ml_pipe = MLPipeline(
    target='no_fragmentation',
    estimator=estimator,
    model_postfix=model_postfix,
    add_smote=add_smote,
)

ml_pipe.run()
```

FINDING: The metrics on the holdout sample worsened with the addition of SMOTE, both for splashing and for no_fragmentation. The CV metrics improved only for no_fragmentation, see `modelling_df4_dimless_smote_2_pure_smote.ipynb`.
