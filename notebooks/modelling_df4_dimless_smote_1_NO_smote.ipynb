# Logit

Splashing. Without SMOTE

```python
estimator = StatsModelsEstimator
target = 'splashing'
estimator_params = {
    'model_class': Logit,
}

ml_pipe = MLPipeline(
    target=target,
    estimator=estimator,
    estimator_params=estimator_params,
    model_postfix=model_postfix,
    add_smote=add_smote,
)

ml_pipe.run()
```

No fragmentation. without SMOTE

```python
estimator = StatsModelsEstimator
target = 'no_fragmentation'
estimator_params = {
    'model_class': Logit,
}

ml_pipe = MLPipeline(
    target=target,
    estimator=estimator,
    estimator_params=estimator_params,
    model_postfix=model_postfix,
    add_smote=add_smote,
)

ml_pipe.run()
```

FINDING: The metrics on the holdout sample worsened with the addition of SMOTE, both for splashing and for no_fragmentation. The CV metrics improved only for no_fragmentation, see `modelling_df4_dimless_smote_2_pure_smote.ipynb`.
