# Settings

```python
features_to_drop = (
    'Re', 
    'We', 
    'init_volume_fraction',
    'particle_droplet_diameter_ratio',
    'sedimentation_Re',
    'particle_liquid_density_ratio',
    # 'sedimentation_Stk',
    # 'sign_sedimentation_Re',
    # 'volume_fraction', 
    # 'relative_roughness', 
    # 'inclination',
    # 'wettability',
    # 'K',
)
log_features=(
    'relative_roughness',
    'sedimentation_Stk',
)
minmax_features_splash=(
    # 'relative_roughness',
    'inclination',
    'volume_fraction',
)
passthrough_features_splash=(
    'wettability',
)
minmax_features_no_fragm=(
    # 'relative_roughness',
)
passthrough_features_no_fragm=(
    'wettability',
    'inclination',
    'volume_fraction',
)
add_const=False
model_postfix='Stk'
```

# Logit

```python
estimator = StatsModelsEstimator(Logit)

ml_pipe = MLPipeline(
    target='splashing',
    estimator=estimator,
    features_to_drop=features_to_drop,
    minmax_features=minmax_features_splash,
    passthrough_features=passthrough_features_splash,
    log_features=log_features,
    add_const=add_const,
    model_postfix=model_postfix,
)

ml_pipe.run()
```
