## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.metrics import matthews_corrcoef
from itertools import combinations
```

## Load data

```python
df = pd.read_excel('../data/data_target.xlsx', index_col=0)
df.info()
```

```python
target_clms = df.columns.drop(['Test #', 'Total number of dropped drops'])
target_clms
```

## Correlation

```python
sns.heatmap(df[target_clms].corr(method='spearman'), annot=True)
```

```python
matthews_corrcoef_dt = {}
for pair in combinations(target_clms, 2):
    matthews_corrcoef_dt[pair] = matthews_corrcoef(df[pair[0]], df[pair[1]])
matthews_corrcoef_dt
```
