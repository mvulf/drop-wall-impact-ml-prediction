Image for experiments visualisation

```python
from matplotlib import pyplot as plt
from PIL import Image
import seaborn as sns
```

```python
path_top = '../results/camera_views/Snapshot_top_v3.jpg'
path_front = '../results/camera_views/Snapshot_front.jpg'
```

```python
# Set the size of the figure
sns.set_style('darkgrid', {'font.family':'serif', 'font.serif':'Times New Roman'})
fig, axs = plt.subplots(2, gridspec_kw={'height_ratios': [0.345, 0.7]},
                        figsize=(4, 4), dpi=500)
axs[0].set_title('(a)')
axs[0].axis('off')
axs[0].imshow(Image.open(path_front))

axs[1].set_title('(b)')
axs[1].axis('off')
axs[1].imshow(Image.open(path_top))
fig.savefig('../results/camera_views/both_cameras_views.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
