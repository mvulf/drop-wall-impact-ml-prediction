Visualisation of loss for NNs

# Libraries

```python
import pandas as pd
from pathlib import Path
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data load

```python
path_xlsxs = Path('../results/nn_logs')
df_splashing_mlp = pd.read_excel(path_xlsxs / 'splashing_mlp.xlsx')
df_no_fragmentation_mlp = pd.read_excel(path_xlsxs / 'no_fragmentation_mlp.xlsx')

df_splashing_tabnet = pd.read_excel(path_xlsxs / 'splashing_tabnet.xlsx')
df_no_fragmentation_tabnet = pd.read_excel(path_xlsxs / 'no_fragmentation_tabnet.xlsx')
```

```python
df_splashing_mlp = df_splashing_mlp[df_splashing_mlp['folder']!='version_7']
df_no_fragmentation_mlp = df_no_fragmentation_mlp[df_no_fragmentation_mlp['folder']!='version_7']

df_splashing_tabnet = df_splashing_tabnet[df_splashing_tabnet['folder']!='version_7']
df_no_fragmentation_tabnet = df_no_fragmentation_tabnet[df_no_fragmentation_tabnet['folder']!='version_7']
```

# Visualisation

```python
import numpy as np
```

```python
sns.set_style('whitegrid', {'font.family':'serif', 'font.serif':'Times New Roman'})
plt.rcParams.update({'font.size': 32})

# Create a figure with 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 12), dpi=600)

# Splashing, MLP
df_splashing_mlp = df_splashing_mlp[df_splashing_mlp['step']<=40]
sns.lineplot(data=df_splashing_mlp, x='step', y='train_loss', marker=None, ci='sd', label='Train', color='blue', ax=axes[0, 0])
sns.lineplot(data=df_splashing_mlp, x='step', y='valid_loss', marker=None, ci='sd', label='Valid', color='orange', ax=axes[0, 0])
axes[0, 0].set_ylabel('Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_title('(a)')
axes[0, 0].legend()
axes[0, 0].set_ylim(bottom=0)
axes[0, 0].xaxis.set_ticks(np.arange(0, 40, 5))
axes[0, 0].grid(True)
xticks = axes[0, 0].get_xticks()
axes[0, 0].set_xticks(xticks[0::2])
axes[0, 0].grid(which='both', linestyle='--', linewidth=0.5)
axes[0, 0].xaxis.set_minor_locator(plt.MultipleLocator(5))  # Adjust the minor locator as needed
axes[0, 0].grid(which='minor', linestyle=':', linewidth=0.5, color=(0.5, 0.5, 0.5))


# Plot 2: Splashing, TabNET
df_splashing_tabnet = df_splashing_tabnet[df_splashing_tabnet['step']<=360]
sns.lineplot(data=df_splashing_tabnet, x='step', y='train_loss', marker=None, ci='sd', label='Train', color='blue', ax=axes[0, 1])
sns.lineplot(data=df_splashing_tabnet, x='step', y='valid_loss', marker=None, ci='sd', label='Valid', color='orange', ax=axes[0, 1])
axes[0, 1].set_ylabel('Loss')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_title('(b)')
axes[0, 1].legend()
axes[0, 1].set_ylim(bottom=0)
axes[0, 1].xaxis.set_ticks(np.arange(0, 361, 50))
axes[0, 1].grid(True)
xticks = axes[0, 1].get_xticks()
axes[0, 1].set_xticks(xticks[0::2])
axes[0, 1].grid(which='both', linestyle='--', linewidth=0.5)
axes[0, 1].xaxis.set_minor_locator(plt.MultipleLocator(50))  # Adjust the minor locator as needed
axes[0, 1].grid(which='minor', linestyle=':', linewidth=0.5, color=(0.5, 0.5, 0.5))

# Plot 3: No Fragmentation, MLP
df_no_fragmentation_mlp = df_no_fragmentation_mlp[df_no_fragmentation_mlp['step']<=35]
sns.lineplot(data=df_no_fragmentation_mlp, x='step', y='train_loss', marker=None, ci='sd', label='Train', color='blue', ax=axes[1, 0])
sns.lineplot(data=df_no_fragmentation_mlp, x='step', y='valid_loss', marker=None, ci='sd', label='Valid', color='orange', ax=axes[1, 0])
axes[1, 0].set_ylabel('Loss')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_title('(c)')
axes[1, 0].legend()
axes[1, 0].set_ylim(bottom=0)
axes[1, 0].yaxis.set_ticks(np.arange(0, 1.5, 0.25))
axes[1, 0].grid(True)
yticks = axes[1, 0].get_yticks()
axes[1, 0].set_yticks(yticks[0::2])
axes[1, 0].grid(which='both', linestyle='--', linewidth=0.5)
axes[1, 0].yaxis.set_minor_locator(plt.MultipleLocator(0.25))  # Adjust the minor locator as needed
axes[1, 0].grid(which='minor', linestyle=':', linewidth=0.5, color=(0.5, 0.5, 0.5))

axes[1, 0].xaxis.set_ticks(np.arange(0, 40, 5))
xticks = axes[1, 0].get_xticks()
axes[1, 0].set_xticks(xticks[0::2])
axes[1, 0].grid(which='both', linestyle='--', linewidth=0.5)
axes[1, 0].xaxis.set_minor_locator(plt.MultipleLocator(5))  # Adjust the minor locator as needed
axes[1, 0].grid(which='minor', linestyle=':', linewidth=0.5, color=(0.5, 0.5, 0.5))

# Plot 4: No Fragmentation, TabNET
df_no_fragmentation_tabnet = df_no_fragmentation_tabnet[df_no_fragmentation_tabnet['step']<=275]
sns.lineplot(data=df_no_fragmentation_tabnet, x='step', y='train_loss', marker=None, ci='sd', label='Train', color='blue', ax=axes[1, 1])
sns.lineplot(data=df_no_fragmentation_tabnet, x='step', y='valid_loss', marker=None, ci='sd', label='Valid', color='orange', ax=axes[1, 1])
axes[1, 1].set_ylabel('Loss')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_title('(d)')
axes[1, 1].legend()
axes[1, 1].set_ylim(bottom=0)
axes[1, 1].xaxis.set_ticks(np.arange(0, 276, 50))
axes[1, 1].grid(True)
xticks = axes[1, 1].get_xticks()
axes[1, 1].set_xticks(xticks[0::2])
axes[1, 1].grid(which='both', linestyle='--', linewidth=0.5)
axes[1, 1].xaxis.set_minor_locator(plt.MultipleLocator(50))  # Adjust the minor locator as needed
axes[1, 1].grid(which='minor', linestyle=':', linewidth=0.5, color=(0.5, 0.5, 0.5))

# Adjust layout
plt.tight_layout()

fig.savefig('../results/losses_nns.pdf', format='pdf', bbox_inches='tight')
# Show the plot
plt.show()
```
