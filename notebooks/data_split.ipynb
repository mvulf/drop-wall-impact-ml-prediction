# Libraries

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
# fixed random seed
RANDOM_STATE = 42
```

# EDA

```python
path_data = Path('../data')
```

```python
df_main = pd.read_excel(path_data / 'df_main.xlsx')
df_modelling_dimensionless = pd.read_excel(path_data / 'df_modelling_dimensionless.xlsx')
df_modelling_full = pd.read_excel(path_data / 'df_modelling_full.xlsx')
df_modelling_no_multicollinearity = pd.read_excel(
    path_data / 'df_modelling_no_multicollinearity.xlsx')
```

```python
common_columns = list(set(df_modelling_dimensionless.columns) & set(df_modelling_full.columns) &\
                set(df_modelling_no_multicollinearity.columns))
print(f'Common columns for dataframes:\n{common_columns}')
```

```python
df_modelling_dimensionless.shape,  df_modelling_full.shape, df_modelling_no_multicollinearity.shape
```

```python
(df_modelling_dimensionless[common_columns] == df_modelling_full[common_columns]).sum()
```

```python
(df_modelling_no_multicollinearity[common_columns] == df_modelling_full[common_columns]).sum()
```

# Train/test split

## Splashing

```python
train_splashing, test_splashing = train_test_split(df_modelling_full, 
                                               test_size=0.2, random_state=RANDOM_STATE,
                                               stratify=df_modelling_full['splashing'])
```

```python
train_splashing['splashing'].value_counts()
```

```python
test_splashing['splashing'].value_counts()
```

```python
train_indexes = train_splashing.index 
test_indexes = test_splashing.index
df_split = pd.DataFrame({
    'index': list(train_indexes)+list(test_indexes),
    'sample': (['train'] * len(train_indexes)) + (['test'] * len(test_indexes))
})
df_split.sample(5, random_state=RANDOM_STATE).T
```

```python
df_split.to_excel(path_data / 'df_ml_split_splashing.xlsx')
```

## Net impact

```python
train_net_impact, test_net_impact = train_test_split(df_modelling_full, 
                                               test_size=0.2, random_state=RANDOM_STATE,
                                               stratify=df_modelling_full['net_impact'])
```

```python
train_net_impact['net_impact'].value_counts()
```

```python
test_net_impact['net_impact'].value_counts()
```

```python
train_indexes = train_net_impact.index 
test_indexes = test_net_impact.index
df_split = pd.DataFrame({
    'index': list(train_indexes)+list(test_indexes),
    'sample': (['train'] * len(train_indexes)) + (['test'] * len(test_indexes))
})
df_split.sample(5, random_state=RANDOM_STATE).T
```

```python
df_split.to_excel(path_data / 'df_ml_split_net_impact.xlsx')
```
