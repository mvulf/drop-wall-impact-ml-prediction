# Libraries

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
# fixed random seed
RANDOM_STATE = 42
```

# EDA

```python
path_data = Path('../data')
```

```python
df_main = pd.read_excel(path_data / 'df_main.xlsx')
df_modelling_dimensionless = pd.read_excel(path_data / 'df_modelling_dimensionless.xlsx')
df_modelling_full = pd.read_excel(path_data / 'df_modelling_full.xlsx')
df_modelling_no_multicollinearity = pd.read_excel(
    path_data / 'df_modelling_no_multicollinearity.xlsx')
```

```python
common_columns = list(set(df_modelling_dimensionless.columns) & set(df_modelling_full.columns) &\
                set(df_modelling_no_multicollinearity.columns))
print(f'Common columns for dataframes:\n{common_columns}')
```

```python
df_modelling_dimensionless.shape,  df_modelling_full.shape, df_modelling_no_multicollinearity.shape
```

```python
(df_modelling_dimensionless[common_columns] == df_modelling_full[common_columns]).sum()
```

```python
(df_modelling_no_multicollinearity[common_columns] == df_modelling_full[common_columns]).sum()
```

Number of experiment (test) matches number of row in every modelling dataframe

# Train/test split

```python
train_indexes, test_indexes = train_test_split(df_modelling_full.index, 
                                               test_size=0.2, random_state=RANDOM_STATE)
```

```python
df_split = pd.DataFrame({
    'index': list(train_indexes)+list(test_indexes),
    'sample': (['train'] * len(train_indexes)) + (['test'] * len(test_indexes))
})
df_split.sample(5, random_state=RANDOM_STATE)
```

```python
df_split.to_excel(path_data / 'df_ml_split.xlsx')
```
