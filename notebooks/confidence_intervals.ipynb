# Splashing

../results/models_modelling_2/logisticregression_splashing_df_modelling_dimensionless_onehot

```python
target = 'splashing'
train, test = get_train_test(
    dataset_filename='df_modelling_dimensionless',
    target=target)
train = train[dict_num_features['df_modelling_dimensionless']+['wettability']+[target]]
test = test[dict_num_features['df_modelling_dimensionless']+['wettability']+[target]]
```

```python
path_pipeline = r'../results/models_modelling_2/logisticregression_splashing_df_modelling_dimensionless_onehot'
pipeline = joblib.load(path_pipeline)
```

```python
column_transformer = pipeline.steps[0][1].fit(train.drop(columns=[target]))
```

## With intercept

```python
X = column_transformer.transform(
    train.drop(columns=[target]))
X = sm.add_constant(X)
y = train[target]
```

```python
model_splashing_intercept = sm.Logit(y, X)
result = model_splashing_intercept.fit()
model_splashing_intercept.exog_names[:] = ['const'] + [
    x.replace('num__', '').replace('cat__', '') for x in column_transformer.get_feature_names_out()]
```

```python
result.summary()
```

```python
result.summary().as_latex()
```

## Without intercept

```python
X = column_transformer.transform(
    train.drop(columns=[target]))
y = train[target]
```

```python
model_splashing = sm.Logit(y, X)
result = model_splashing.fit()
model_splashing.exog_names[:] = [
    x.replace('num__', '').replace('cat__', '') for x in column_transformer.get_feature_names_out()]
```

```python
result.summary()
```

```python
result.summary().as_latex()
```
