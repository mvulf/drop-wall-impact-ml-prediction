# Data load

```python
df_metrics = pd.read_excel('../results/metrics_modelling4.xlsx')
df_stump = pd.read_excel('../results/metrics_modelling3.xlsx').iloc[-2:]
df_metrics = pd.concat((df_metrics, df_stump)).reset_index(drop=True)
```

# Update data

```python
df_metrics = df_metrics.loc[
    df_metrics['model'].apply(lambda x:
                              x in [
        'DecisionStump_splashing_greater_or_equal_kl',
        'Logit_splashing_default',
        'CatBoostClassifier_splashing_smote_opt-model'
        ])][[
            'model', 'holdout_test_accuracy',
                      'holdout_test_f1_macro', 'holdout_test_roc_auc'
        ]].sort_values(by=['holdout_test_f1_macro']).reset_index(drop=True)
df_metrics
```

```python
first_model = df_metrics.iloc[0]

for metric in ['holdout_test_accuracy', 'holdout_test_f1_macro',
       'holdout_test_roc_auc']:
    df_metrics[f'{metric}_pct_change'] = ((df_metrics[metric] - first_model[metric]) / first_model[metric]) * 100
    df_metrics.loc[0, f'{metric}_pct_change'] = 0
df_metrics
```
